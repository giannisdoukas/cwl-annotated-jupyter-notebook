## Generate fake data

```python
import dask.dataframe as dd
```

```python
df = dd.demo.make_timeseries(start='2000', end='2015', dtypes={'A': float, 'B': int},
                             freq='5s', partition_freq='3M', seed=1234)
```

## Compute and plot a cumulative sum

```python
fig = df.A.cumsum().resample('1w').mean().compute().plot();
```

```python
from IPython.display import FileLink, FileLinks
data_img: 'CWLFilePathOutput' = 'data.png'
print(f'saving output image to file {data_img}')
fig.figure.savefig(data_img)
FileLink(data_img)
```
